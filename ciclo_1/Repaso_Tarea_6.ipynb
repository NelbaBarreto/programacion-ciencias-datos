```python
import pandas as pd
import numpy as np
import seaborn as sns
import random

# Cargar datos de Titanic desde seaborn
df = sns.load_dataset('titanic')
print(df.shape)

df
```

```python
# Describe el dataframe, sus columnas y el tipo de dato de cada una. (usar método info() y describe()
df.info()
```

```python
df.describe()
```

```python
# Crear una tabla pivote para encontrar la suma de las tarifas (fare) por clase (class).
sum_tarifa_por_clase = pd.pivot_table(df, index="class", values="fare", aggfunc="sum")
print("Suma de las tarifas (fare) por clase (class)", sum_tarifa_por_clase, sep="\n")
```

```python
# Crear una tabla pivote para encontrar el promedio de las tarifas (fare) por clase (class).
avg_tarifa_por_clase = pd.pivot_table(df, index="class", values="fare", aggfunc="mean")
print("Promedio de las tarifas (fare) por clase (class)", avg_tarifa_por_clase, sep="\n")
```

```python
# Crear una tabla pivote para encontrar la suma y el promedio de las tarifas (fare)
# por clase (class).
agg_tarifa_por_clase = pd.pivot_table(df, index="class", values="fare", aggfunc=["sum", "mean"])
print("Suma y promedio de las tarifas (fare) por clase (class)", agg_tarifa_por_clase, sep="\n")
```

```python
# Crear categorías de tarifas (fare) usando qcut
df["categorias_tarifas"] = pd.cut(df["fare"], 4, labels=["bajo", "medio-bajo", "medio-alto", "alto"])

# crear una tabla pivote basada en esas categorías, mostrando la cantidad
# de pasajeros para esa categoría.
categorias_por_tarifa = pd.pivot_table(df, index="categorias_tarifas", values="fare", aggfunc="count")
print("Cantidad de pasajeros por categoría de tarifas", categorias_por_tarifa, sep="\n")
```

```python
### Ejercicio 6 y 7
first_names=('Horacio', 'Margarita', 'Pedro','Juan','Andrea','Jose')
last_names=('Martinez','Perez','Peña', 'Soto', 'Ruiz', 'Ferreira')

group = []
for i in range(891):
    full_name=random.choice(first_names)+" "+random.choice(last_names)
    group.append(full_name)

# Antes de anexar al dataframe, deben modificar la lista para que contenga Mrs
# para los nombres femeninos y Mr para los nombres masculinos. La lista final solo
# debe contener el título y el apellido del pasajero. Esta lista pasará a ser
# la columna name del dataframe y se harán sobre esta columna, los ejercicios 5 y 6
def final_name(x):
  first_name = x.split()[0]
  last_name = x.split()[1]

  if first_name in ["Margarita", "Andrea"]:
    title = "Mrs"
  else:
    title = "Mr"

  return title + " " + last_name

group = list(map(final_name, group))

df["name"] = group

df.head()
```

```python
# Dividir la columna de nombres (name) en título y apellido.
df[["titulo", "apellido"]] = df["name"].str.split(expand=True)

print(df[["name", "titulo", "apellido"]].head())
```

```python
# Filtrar filas donde los nombres (name) contienen "Mrs".
print("Filas que contienen Mrs", df[df["name"].str.contains("Mrs", case=False)], sep="\n")
```

```python
# Para este ejercicio crearemos una columna ficticia de fechas (fecha de embarque).
# Usar pd_date_range para crear columna ficticia para cada fila desde '1912-04-10'
# (es decir, '1912-04-10' para la fila 1, '1912-04-11' para la fila 2, etc).
date_range = pd.date_range(start="1912-04-10", periods=len(df), freq='D')
df["fecha_embarque"] = date_range

df["fecha_embarque"].head()
```

```python
# Cambiar el formato de una columna de fechas a al formato día, mes, año.
print(df["fecha_embarque"].dt.strftime("%d-%m-%Y"))
```

```python
# Agrupar por fecha de embarque y calcular el número de sobrevivientes (survived)
# por mes, su promedio de edad y cantidad de hombres y mujeres.
df["mes"] = df["fecha_embarque"].dt.month
df[df.survived == 1].groupby("mes").agg(
    num_survivors=("survived", "sum"),
    avg_age=("age", "mean"),
    num_males=("sex", lambda x: (x == "male").sum()),
    num_females=("sex", lambda x: (x == "female").sum())
)
```
