Este conjunto de datos proporciona información sobre los pasajeros del Titanic, como su clase, edad, género, tarifa pagada, etc. Los ejercicios cubrirán las operaciones de rename, merge, concat, agregaciones y groupby en pandas. # Ejercicio 1

Importa el dataset titanic. Muestre las 10 primeras y últimas filas.

Renombra las columnas 'age' a 'edad' y 'fare' a 'tarifa' en el dataset titanic.

```python
import pandas as pd
import seaborn as sns

# Cargar dataset Titanic de seaborn
titanic = sns.load_dataset('titanic')

# Mostrar las 10 primeras filas del dataset
titanic.head()
```

```python
# Mostrar las 10 últimas filas del dataset
titanic.tail()
```

```python
# Renombra las columnas 'age' a 'edad' y 'fare' a 'tarifa'.
titanic.rename(columns={"age":"edad", "fare":"tarifa"}, inplace=True)

titanic.head()
```

# Ejercicio 2

Con el dataframe adicional con información de clases y costos promedio, y haz un merge (inner) con el dataframe titanic basado en la columna 'class'.

Mostrar las 10 primeras filas.

```python
# Dataframe adicional
costos = pd.DataFrame({
    'class': ['First', 'Second', 'Third'],
    'costo_promedio': [84.154687, 20.662183, 13.675550]
})

# Haz un merge (inner) con el dataframe titanic basado en la columna 'class'.
titanic = titanic.merge(costos, on="class")
print(titanic.head())
```

# Ejercicio 4

Calcula el promedio de edad y tarifa por cada clase en el dataset titanic.

```python
promedio_edad_tarifa = titanic.groupby("class")[["edad", "tarifa"]].mean()

print("Promedio de edad y tarifa por cada clase:", promedio_edad_tarifa, sep="\n")
```

# Ejercicio 5

Usa el método aggregate para calcular el mínimo y máximo de edad y tarifa para cada clase en el dataset titanic.

```python
min_max_edad_tarifa = titanic.groupby("class")[["edad", "tarifa"]].aggregate(["min", "max"])

print("Mínimo y máximo de edad y tarifa por cada clase:", min_max_edad_tarifa, sep="\n")
```

# Ejercicio 6
Filtra los pasajeros que tienen más de 30 años y pertenecen a la clase 'First'.

```python
print("Los pasajeros con más de 30 años que pertenecen a la clase First son:", titanic[(titanic["edad"] > 30) & (titanic["class"] == "First")], sep="\n")
```

# Ejercicio 7

Transforma la columna 'age' para normalizarla restando la media y dividiendo por la desviación estándar.

```python
def normalizar(x):
  return (x - x.mean()) / x.std()

titanic["edad_normalizada"] = titanic["edad"].transform(normalizar)

print(titanic.head())
```

# Ejercicio 8

Aplica una función personalizada que clasifique la edad en 'niño', 'adulto', 'anciano' en una nueva columna.

```python
def classify_by_age(x):
  if x < 18:
    clasificacion = "niño"
  elif x < 68:
    clasificacion = "adulto"
  else:
    clasificacion = "anciano"

  return clasificacion

titanic["clasificacion_edad"] = titanic["edad"].apply(classify_by_age)

print(titanic.head())
```

# Ejercicio 9

Filtra los pasajeros que sobrevivieron y calcula la tarifa promedio por clase para estos pasajeros.

```python
tarifa_promedio_sob = titanic[titanic["survived"] == 1].groupby("class")["tarifa"].mean()

print("La tarifa promedio por clase para los pasajeros que sobrevivieron fue de:", tarifa_promedio_sob, sep="\n")
```

# Ejercicio 10

Crea un nuevo dataframe que contenga solo los pasajeros con más de 30 años, luego agrupa por clase y sexo, y calcula el promedio de edad y tarifa.



```python
# Crea un nuevo dataframe que contenga solo los pasajeros con más de 30 años
titanic_nuevo = titanic[titanic["edad"] > 30]

group_by_class_sex = titanic_nuevo.groupby(["class", "sex"])
res = group_by_class_sex[["edad", "tarifa"]].mean()

print("Promedio de edad y tarifa para pasajeros con más de 30 años agrupados por clase y sexo:", res, sep="\n")
```
