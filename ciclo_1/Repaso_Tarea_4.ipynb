# Series
1. Crea una Serie llamada s con los siguientes valores: 10, 20, 30, 40, 50.
2. Imprime la Serie s.
3. Accede al tercer elemento de la Serie s.
4. Modifica el segundo elemento de la Serie s para que sea 25.
5. Calcula la suma de todos los elementos en la Serie s.
6. Elimina el elemento 50.


```python
import pandas as pd
import numpy as np
```

```python
# Crear una Serie llamada s con los siguientes valores: 10, 20, 30, 40, 50.
s = pd.Series([10, 20, 30, 40, 50])

# Imprime la Serie s.
print("Serie s:", s, sep= "\n")

# Accede al tercer elemento de la Serie s.
print("\nEl tercer elemento de la serie es:", s[2])

# Modifica el segundo elemento de la Serie s para que sea 25.
s[1] = 25

# Calcula la suma de todos los elementos en la Serie s.
print("La suma de todos los elementos de la serie es:", s.sum())

# Elimina el elemento 50.
del s[4]

print("\nSerie s:", s, sep= "\n")
```

# DataFrames
1. Crea un DataFrame llamado df con dos columnas: "A" con valores [1, 2, 3, 4, 5] y "B" con valores [10, 20, 30, 40, 50].
2. Muestra las primeras tres filas del DataFrame df.
3. Accede a la columna "B" del DataFrame df.
4. Calcula la media de la columna "B" en el DataFrame df.
5. Agrega una nueva columna llamada "C" al DataFrame df con valores [5, 10, 15, 20, 25]. Usa la función range.

```python
# Crea un DataFrame llamado df con dos columnas: "A" con valores [1, 2, 3, 4, 5] y "B" con valores [10, 20, 30, 40, 50].
data = {"A": [1, 2, 3, 4, 5], "B": [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Muestra las primeras tres filas del DataFrame df.
print("Primeras tres filas del DataFrame", df[0:3], sep="\n")

# Accede a la columna "B" del DataFrame df.
print("\nColumna B del DataFrame", df["B"], sep="\n")

# Calcula la media de la columna "B" en el DataFrame df.
print("\nLa media de la columna B es:", df["B"].mean())

# Agrega una nueva columna llamada "C" al DataFrame df con valores [5, 10, 15, 20, 25].
# Usa la función range.
df["C"] = range(5, 30, 5)
print("\nDataFrame modificado", df, sep="\n")
```

# Index
1. Crea una Serie llamada s con los valores [10, 20, 30, 40, 50] y asigna como índices las letras "a", "b", "c", "d", "e".
2. Imprime el índice de la Serie s.
3. Selecciona y muestra los elementos 'a', 'b' y 'c'.
4. Crea un DataFrame llamado df con dos columnas "A" y "B" y valores de [2, 4, 6, 8, 10] y [100, 200, 300, 400, 500]. Asigna como índices los nombres "Alice", "Bob", "Charlie", "David", "Emma".


```python
# Crea una Serie llamada s con los valores [10, 20, 30, 40, 50] y asigna como índices las letras "a", "b", "c", "d", "e".
s = pd.Series([10, 20, 30, 40, 50], index=["a", "b", "c", "d", "e"])

# Imprime el índice de la Serie s.
print("El índice de la Serie s es:", s.index, sep="\n")

# Selecciona y muestra los elementos 'a', 'b' y 'c'.
print("\nLos elementos a, b y c son:", s.loc["a":"c"], sep="\n")

# Crea un DataFrame llamado df con dos columnas "A" y "B" y valores de [2, 4, 6, 8, 10]
# y [100, 200, 300, 400, 500]. Asigna como índices los nombres "Alice", "Bob", "Charlie", "David", "Emma".
data = {"A": [2, 4, 6, 8, 10], "B": [100, 200, 300, 400, 500]}
index = ["Alice", "Bob", "Charlie", "David", "Emma"]

df = pd.DataFrame(data, index)
print("\nDatos del nuevo DataFrame:", df, sep="\n")
```

# Indexación y selección de datos
1. Selecciona la segunda fila del DataFrame df.
2. Selecciona la columna "A" del DataFrame df.
3. Selecciona las filas con índices "Alice" y "Charlie" del DataFrame df.
4. Elimina la columna 'A' y la fila 'Alice'. Imprime el dataframe resultante.



```python
# Selecciona la segunda fila del DataFrame df.
print("Segunda fila del DataFrame:", df.iloc[1], sep="\n")

# Selecciona la columna "A" del DataFrame df.
print("\nColumna A del DataFrame:", df["A"], sep="\n")

# Selecciona las filas con índices "Alice" y "Charlie" del DataFrame df.
print("\nFilas Alice y Charlie del DataFrame:", df.loc[["Alice", "Charlie"]], sep="\n")

# Elimina la columna 'A' y la fila 'Alice'. Imprime el dataframe resultante.
df.drop(axis="columns", columns=["A"], inplace=True)
df.drop("Alice", inplace=True)

print("\nDataFrame luego de eliminar la columna A y la fila Alice:", df, sep="\n")
```

# Manejo de datos faltantes

2. Elimina todas las filas con valores nulos en el DataFrame df.

```python
df = pd.DataFrame([[1, np.nan, np.nan], [2, 3, np.nan], [np.nan, 4, 6]])
print("\nDataFrame original", df, sep="\n")

# Dada la matriz df, reemplaza todos los valores nulos en el DataFrame df con ceros.
print("\nDataFrame con valores nulos reemplazados con ceros", df.fillna(0), sep="\n")

# Elimina todas las filas con valores nulos en el DataFrame df.
print("\nDataFrame con valores nulos eliminados", df.dropna(), sep="\n")
```

# Tipos de valores nulos

1. Verifica si hay valores nulos en el DataFrame df.
2. Cuenta el número de valores nulos en cada columna del DataFrame df.



```python
# Verifica si hay valores nulos en el DataFrame df.
print("Valores nulos en el DataFrame", df.isnull(), sep="\n")

# Cuenta el número de valores nulos en cada columna del DataFrame df.
print("\nValores nulos en el DataFrame por columna", df.isnull().sum(), sep="\n")
```
